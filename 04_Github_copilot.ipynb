```python
# how to make plots in matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

df= sns.load_dataset('tips')
df.head()
```

```python
#How to make lineplot
sns.lineplot(x='day', y='total_bill', data=df)
```

```python
#How to make barplot
sns.barplot(x='day', y='total_bill', data=df)
```

```python
#How to make boxplot
sns.boxplot(x='day', y='total_bill', data=df)
```

```python
#How to make boxenplot
sns.boxenplot(x='day', y='total_bill', data=df)
```

```python
#How to make catplot
sns.catplot(x='day', y='total_bill', data=df)
```

```python
#How to make distplot
sns.distplot(df['total_bill'])
```

```python
#How to make residplot
sns.residplot(df['total_bill'], df['tip'])
```

```python
#How to make lmplot
sns.lmplot(x='total_bill', y='tip', data=df)
```

```python
#How to make heatmap
sns.heatmap(df.corr(), annot=True)
```

```python
#How to make pairplot
sns.pairplot(df)
```

```python
#How to make factorplot
sns.factorplot(x='day', y='total_bill', data=df)
```
